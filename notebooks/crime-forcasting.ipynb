```python

crime["date"] = pd.to_datetime(crime["Year"].astype(str) + crime["Month"], format="%Y%B")
crime.set_index("date",inplace=True)
numeric_columns=crime.select_dtypes(include="number")
monthly_data=numeric_columns.resample("M").sum()
```

```python
monthly_data.head()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(monthly_data['Victim Count'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

# If p-value is greater than 0.05, we proceed with differencing.
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plot_acf(monthly_data['Victim Count'].dropna(), lags=20)
plot_pacf(monthly_data['Victim Count'].dropna(), lags=20)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(monthly_data['Victim Count'], order=(1,0,1))
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())
```

```python
# Forecasting the next 12 months
forecast = model_fit.get_forecast(steps=24)
forecast_conf = forecast.conf_int()

plt.figure(figsize=(12,6))
plt.plot(monthly_data['Victim Count'], label='Historical')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(forecast_conf.index, forecast_conf.iloc[:, 0], forecast_conf.iloc[:, 1], color='pink')
plt.title('Forecast vs Actuals')
plt.legend()
plt.show()
```
